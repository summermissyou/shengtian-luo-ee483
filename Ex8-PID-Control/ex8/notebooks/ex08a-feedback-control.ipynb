Let us review some control theory concepts:

<h2 style= "color:coral"> Dynamic Systems</h2>

- Equations of motion describe movement by summing forces
- We can determine the net force on an rigid body and provide a control signal to modify it
$$F = ma = m\ddot{x}$$

Usually, we rewrite it as:

$$\ddot{x} = \frac{\sum F_i}{m}$$ 

where $F_i$ are the forces applied to the rigid body.  
For example, the forces in the MM are:
- Left motor force (at the wheel contact with the surface)
- Right motor force (at the wheel contact with the surface)
- Friction forces (opposite to motor forces)
- Drag

That is the bulk of it.  

Actually, we can model the MM dynamics in this manner.   
The dynamics is for the position of the MM $(x,y,\theta)$

$$\ddot{x} = v_y\dot{\theta}+\frac{F_{x,L}+F_{x,R}}{m}$$
$$\ddot{y} = -v_x\dot{\theta}+\frac{F_{y,L}+F_{y,R}}{m}$$
$$\ddot{\theta} = \frac{L}{J}(F_{x,R}-F_{x,L}) -\frac{c}{J}(F_{y,R}+F_{y,L})$$

This is a somewhat messy (nonlinear and coupled) but overall good model for the MM.  
With some assumptions and simplications we can get a adequate MM model.   

$$\ddot{x} = v\cos(\theta)$$
$$\ddot{y} =  v\sin(\theta)$$
$$\ddot{\theta} = \omega$$

where $v$ is the linear velocity of the robot and $\omega$ is the angular velocity.  
This is the dynamics we are using to simulate the MM in this exercise.

<h2 style= "color:coral"> Simple types of controllers</h2>

<h3 style= "color:red"> Bang-bang controllers</h3>

- An "on"/"off" type of controller. 
- No in between, e.g., motor max speed/motor off

A good example of this type of controller is in our house: Heating system
- When it is cold: Heater on
- When it is hot: Heater off

Usually, it is implemented with hysteresis 

<br><img src="../assets/ex8-imgs/feedback-hysteresis.png" style="width: 20%; height: auto;"></br>

- What if we need something in on and off?

<h3 style= "color:red"> Proportional controllers</h3>

- Uses a proportional constant times the error as a control signal.  
$$control\_signal = constant * (desired\_temp - current\_temp)$$
- Need to tune this constant

MM Example: Duckie moves straight forward and need to stop at a red line.  
With proportional controller, it will:
- Drive straight at a speed proportional to the distance from the line
- Tune the constant until results are acceptable 

$$control\_signal = constant*error$$
Where $error$ is the distance left to reach the red line.
