Robots motions can be broken mathematically into two types:
1. Rotations
2. Translations

<h2 style= "color:coral">Robot Rotations
</h2>


Movement of the robot around its center point.  
In our case, rotation with respect to the center of mass

<br><img src="../assets/ex6-imgs/transf-rotation.gif" style="width: 25%; height: auto;"></br>

Rotations transformations are capture by a rotation matrix. 

In 2D we rotate with respect to a $z$ coordinate that it is orthogonal to the $(x,y)$ plane

$$
R = \begin{bmatrix} 
\cos(\theta) & -\sin(\theta)\\
\sin(\theta) & \cos(\theta)
\end{bmatrix}$$

The angle $\theta$ is the roration angle. 
In the video above, the values of theta are:
1. $\theta = -45^{\circ}$   
$
R = \begin{bmatrix} 
0.707 & -0.707\\
0.707 & 0.707
\end{bmatrix}$

2. $\theta =  90^{\circ}$  
$
R = \begin{bmatrix} 
0 & -1\\
1& 0
\end{bmatrix}$

Let's assume that the initial position of the top of the triangle is $(1,0)$

After the first rotation, the position is:
1. $ \begin{bmatrix} 
x_1\\
y_1
\end{bmatrix}= \begin{bmatrix} 
0.707 & 0.707\\
-0.707 & 0.707
\end{bmatrix}
\begin{bmatrix} 
1\\
0
\end{bmatrix} = \begin{bmatrix} 
0.707\\
-0.707
\end{bmatrix}$

Then, after the second rotation it is:

2. $ \begin{bmatrix} 
x_2\\
y_2
\end{bmatrix}= \begin{bmatrix} 
0 & -1\\
1 & 0
\end{bmatrix}
\begin{bmatrix} 
0.707\\
-0.7070
\end{bmatrix} = \begin{bmatrix} 
0.707\\
0.707
\end{bmatrix}$


In 3D, we have three rotational matrices.  
One for each rotation angle $(\phi, \psi, \theta)$

<h2 style= "color:coral">Robot Translations
</h2>

Movement of the robot through the coordinate frame on a given direction  
Usually, the direction is the direction the robot is facing: forward or backward

<br><img src="../assets/ex6-imgs/transf-transl.gif" style="width: 25%; height: auto;"></br>

Mathematically, every point of the robot is translated by a given amount $\Delta$.  
In matrix notations for 2D the new robot position is:

$$
\begin{bmatrix} 
x_{new}\\
y_{new}
\end{bmatrix} = \begin{bmatrix} 
x_{old}\\
y_{old}
\end{bmatrix} +
\begin{bmatrix} 
\Delta_x\\
\Delta_y
\end{bmatrix}
$$

Let's assume that the initial position of the top of the triangle is $(0.707,0.707)$.   
The robot moves one unit in the current direction of $45^{\circ}$: $\Delta = (0.707,0.707)$   

$$
\begin{bmatrix} 
x_{3}\\
y_{3}
\end{bmatrix} =  
\begin{bmatrix} 
0.707\\
0.707
\end{bmatrix} +
\begin{bmatrix} 
0.707\\
0.707
\end{bmatrix}
=
\begin{bmatrix} 
1.414\\
1.414
\end{bmatrix} 
$$


<h2 style= "color:coral">Rotations and Translations
</h2>




Rotations and translations can be combined in one matrix that contains all rigid motions of the robot.  
These special matrices are called **Special Euclidean (SE) Group** matrices.  
They are formed as:

$$
T = \begin{bmatrix} 
R & p\\
0 & 1
\end{bmatrix}
$$

Here $R$ is a rotation matrix while $p$ is the current position of your robot. 


In 2D, called SE(2) matrix:

$$
T = \begin{bmatrix} 
\cos(\theta) & -\sin(\theta) & x\\
\sin(\theta) & \cos(\theta) & y\\
0 & 0 & 1
\end{bmatrix}$$

In our context, we have the world frame and the robot frame. 

<br><img src="../assets/ex6-imgs/transf-positions.png" style="width: 25%; height: auto;"></br>

Herein, we use:

- World frame: $(x_w,y_w)$
- Robot frame: $(x_r,y_r)$

Given that the robot is in position $(x^r_w = 0, x^r_w = 0)$ in the world coordinate.   
Next, the robot moves:  
- Rotates by $45^{\circ}$, $\theta = 45^{\circ}$
- Forward in the $x$ direction of the robot frame by 1 unit -  $\Delta^{x}_{r} = 1$ and $\Delta^{y}_r = 0$

**REMARK**
*Rotation is always performed first when using SE(2).*

To calculate a new position of the robot in the world frame with 2D transformation, we have


$$
\begin{bmatrix} 
x^{rnew}_{w}\\
y^{rnew}_w\\
1
\end{bmatrix} 
= \begin{bmatrix} 
\cos(\theta) & -\sin(\theta) & x^{r}_{w}\\
\sin(\theta) & \cos(\theta) & y^{r}_{w}\\
0 & 0 & 1
\end{bmatrix}
\begin{bmatrix} 
\Delta^{x}_r\\
\Delta^{y}_r\\
1
\end{bmatrix} $$

$$
\begin{bmatrix} 
x^{rnew}_{w}\\
y^{rnew}_w\\
1
\end{bmatrix} 
= \begin{bmatrix} 
0.707 & -0.707 & 0\\
0.707 & 0.707 & 0\\
0 & 0 & 1
\end{bmatrix}
\begin{bmatrix} 
1\\
0\\
1
\end{bmatrix} $$

$$
\begin{bmatrix} 
x^{rnew}_{w}\\
y^{rnew}_w\\
1
\end{bmatrix} 
= 
\begin{bmatrix} 
0.707\\
0.707\\
1
\end{bmatrix} $$


For 3D, we have similar results but we have three transformation matrices. 
One for each rotation matrix.
