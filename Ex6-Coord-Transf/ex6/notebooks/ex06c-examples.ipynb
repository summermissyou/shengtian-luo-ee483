<h2 style= "color:coral">Problem Statement Robot Moving
</h2>

<br><img src="../assets/ex6-imgs/ex-prob-diag2.png" style="width: 35%; height: auto;"></br>


**Note:** Image is not to scale

- Your robot is at position (0,0) facing the x-axis
- It turns 90 degrees left and moves 10 m forward
- It turns 45 degrees right and moves 5 m backward


- First we have to transform the movements in robot frame coordinates to world
- We can split this process up into three parts:
    1. Finding the rotation matrix: $^{W}R_{R}$
    2. Finding the robot position vector in world coordinates: $^{W}p_{old}$
	3. Finding the delta matrix in robot coordinates: $^{R}\Delta$
- Then we combine these results into one transformation matrix: $^{W}T_{R} = \begin{bmatrix} ^{R}R_{S} & ^{R}p_{S} \\ 0 & 1 \end{bmatrix}$
- Next we compute new position $^{W}p_{new} = ^{R}T_{S}\ \ ^{R}\Delta$


- The formula for a 2D rotation matrix is:

$$^{R}R_{S} = 
\begin{bmatrix}
cos(\theta) & -sin(\theta)  \\
sin(\theta) & cos(\theta) 
\end{bmatrix}
$$

- Our theta value is 90 degrees so we get:

$$^{W}R_{R} = 
\begin{bmatrix}
0 & -1  \\
1 & 0 
\end{bmatrix}
$$


- The initial position of the robot is:

$$^{W}P_{old} = \begin{bmatrix}
x_{old} \\
y_{old}
\end{bmatrix}= \begin{bmatrix}
0\\
0
\end{bmatrix}
$$

- The $\Delta$ movements are

$$^{R}\Delta = \begin{bmatrix}
\Delta_x \\
\Delta_y
\end{bmatrix}= \begin{bmatrix}
10\\
0
\end{bmatrix}
$$

- Combining everything, we have

$$
\begin{bmatrix}
^{W}p_{new} \\
1
\end{bmatrix}
=
^{W}T_{R}
\begin{bmatrix}
^{R}\Delta \\
1
\end{bmatrix}
=
\begin{bmatrix}
0 & -1 & 0 \\
1 & 0 & 0 \\
0 & 0 & 1
\end{bmatrix}
\begin{bmatrix}
10 \\
0 \\
1
\end{bmatrix}
=
\begin{bmatrix}
0 \\
10 \\
1
\end{bmatrix}
$$


- Similarly, we can calculate the transformation for the second movement

$$
\begin{bmatrix}
^{W}p_{new} \\
1
\end{bmatrix}
=
\begin{bmatrix}
0.707 & 0.707 & 0 \\
-0.707 & 0.707 & 10 \\
0 & 0 & 1
\end{bmatrix}
\begin{bmatrix}
-5 \\
0 \\
1
\end{bmatrix}
=
\begin{bmatrix}
-3.535\\
6.465 \\
1
\end{bmatrix}
$$

<h2 style= "color:coral">Problem Statement 
</h2>

<br><img src="../assets/ex6-imgs/ex-prob-diag.png" style="width: 35%; height: auto;"></br>

**Note:** Image is not to scale

- Your robot is at position (10,20) facing 45 degrees to the left of the x-axis
- It has a sensor 1 meter forward of the center of the robot facing 90 degrees to the right relative to robot x-axis
- In robot and global coordinates, what is the position of the following obstacles:  
    a. (5,5)  
    b. (10,25)  
    c. (2,6)  

<h2 style= "color:coral">Solution
</h2>

- First we have to transform the sensor coordinates to be in robot frame coordinates
- We can split this process up into two parts:
    1. Finding the rotation matrix: $^{R}R_{S}$
    2. Finding the position matrix: $^{R}p_{S}$
- Then we combine these results into one transformation matrix: $^{R}T_{S} = \begin{bmatrix} ^{R}R_{S} & ^{R}p_{S} \\ 0 & 1 \end{bmatrix}$

#### Rotation Matrix ####

- The formula for a 2D rotation matrix is:

$$^{R}R_{S} = 
\begin{bmatrix}
cos(\theta) & -sin(\theta)  \\
sin(\theta) & cos(\theta) 
\end{bmatrix}
$$

- Our theta value is -90 degrees so we get:

$$^{R}R_{S} = 
\begin{bmatrix}
0 & 1  \\
-1 & 0 
\end{bmatrix}
$$

#### Position Matrix ####
- The formula for a 2D position matrix is:

$$^{R}P_{S} = \begin{bmatrix}
\Delta x  \\
\Delta y
\end{bmatrix}
$$

- Our $\Delta x$ value is 1 and $\Delta y$ is 0:
$$^{R}P_{S} = \begin{bmatrix}
1  \\
0
\end{bmatrix}
$$

#### Combining Rotation and Position Matrix ####

$$^{R}T_{S} = \begin{bmatrix}
^{R}R_{S} & ^{R}p_{S}  \\
0 & 1
\end{bmatrix} = 
\begin{bmatrix}
0 & 1 & 1  \\
-1 & 0 & 0 \\
0 & 0 & 1
\end{bmatrix}
$$

### Finidng Transform for Robot to World Frame ###

- Second we have to find the transformation matrix for the robot to the world frame
#### Rotation Matrix ####

- The 2D rotation matrix with theta equal to 45 degrees is:

$$^{W}R_{R} = 
\begin{bmatrix}
cos(45) & -sin(45)  \\
sin(45) & cos(45) 
\end{bmatrix} = 
\begin{bmatrix}
\frac{1}{\sqrt{2}} & -\frac{1}{\sqrt{2}}  \\
\frac{1}{\sqrt{2}} & \frac{1}{\sqrt{2}}
\end{bmatrix}
$$

#### Position Matrix ####
- The 2D position matrix is:

$$^{W}P_{R} = \begin{bmatrix}
\Delta x  \\
\Delta y
\end{bmatrix} =  \begin{bmatrix}
10  \\
20
\end{bmatrix}
$$

#### Combining Rotation and Position Matrix ####

$$^{W}T_{R} = \begin{bmatrix}
^{W}R_{R} & ^{W}p_{R}  \\
0 & 1
\end{bmatrix} = 
\begin{bmatrix}
\frac{1}{\sqrt{2}} & -\frac{1}{\sqrt{2}} & 10 \\
\frac{1}{\sqrt{2}} & \frac{1}{\sqrt{2}} & 20 \\
0 & 0 & 1
\end{bmatrix}
$$



<h2 style= "color:coral">Applying Transformation Matrices
</h2>

#### Sensed Object in Robot Coordinates ####  
- To find the sensed object in robot coordinates for point a:

$$
\begin{bmatrix}
^{R}p_{a} \\
1
\end{bmatrix}
=
^{R}T_{S}
\begin{bmatrix}
^{S}p_{a} \\
1
\end{bmatrix}
=
\begin{bmatrix}
0 & 1 & 1 \\
-1 & 0 & 0 \\
0 & 0 & 1
\end{bmatrix}
\begin{bmatrix}
5 \\
5 \\
1
\end{bmatrix}
=
\begin{bmatrix}
6 \\
-5 \\
1
\end{bmatrix}
$$
- To find the sensed object in robot coordinates for point b:
$$
\begin{bmatrix}
^{R}p_{b} \\

\end{bmatrix}
=

#### Sensed Object in World Coordinates ####

^{W}T_{R}

- To find the sensed object in world coordinates for point b:
$$
\begin{bmatrix}
^{W}p_{b} \\

^{W}T_{R}
\begin{bmatrix}

- To find the sensed object in world coordinates for point c:
$$

^{W}T_{R}
