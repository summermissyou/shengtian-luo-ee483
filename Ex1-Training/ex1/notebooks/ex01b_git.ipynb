</h2>

### Pulling the most update version
1. Check if you have modified any of the notebooks (Easier to merge if the notebooks are as in the mainstream)  
`git status`

2. Restore any file in the `ex1/notebooks` appear as modified  
`git restore <path to file>` For example, `git restore Ex1-Training/ex1/notebooks/ex01-overview.ipynb`

3. Pull upstream repository   
`git pull upstream main`

4. Merge with vim. Press `Esc` then type `:wq` Enter (it will save your merging)  

5. Push the merging to your repo  
`git push`

All done, your exercise should be up to date.



### Submission
1. Check which files changed  
`git status`

2. Add them to this staged commit  
`git add -A`

3. Make the commit  
`git commit -m 'your message for the commit`

4. Push it to your repo on GitHub
`git push`

5. Tag with based on the exercise or lab. Replace `<tagname> with your the tag` 
`git tag <tagname>`

6. Push the tag to Github  
`git push origin <tagname>`

7. Verify on Github.com that your submission is there, in the correct tag
