<h2 style= "color:coral">Overview
</h2>


The MM is moving around the world.  
Our goal is to break its movements into steps.  
We assume the MM starts in position $(x,y,\theta)$  and moves to final position $(x',y',\theta')$  
Moreover, $2L$ is the distance in between the two wheels of the $MM$  

<br><img src="../assets/ex7-imgs/kin-overview1.png" style="width: 20%; height: auto;"></br>

The change in position is captured by the differences in between initial and final positions.
<br><img src="../assets/ex7-imgs/kin-overview2.png" style="width: 20%; height: auto;"></br>
$$x'= x+\Delta x$$
$$y'= y+\Delta y$$
$$\theta'= \theta+\Delta \theta$$

In notebook 07b, we learned about calculating the distance traveled by each wheel using enconders.   

<h2 style= "color:coral">Using the Encoders Values
</h2>

Overall, the MM will travel a total distance $\Delta s$ as shown below.  
Moreover, it rotates $\alpha$ with respect to its instant center of curvature (ICC).  
For frequent updates, $\alpha \approx  \Delta \theta

<br><img src="../assets/ex7-imgs/kin-encoder1.png" style="width: 20%; height: auto;"></br>

The total distance can be broken into the total distance percurred by each wheel, $\Delta s_L$ for the left wheel and $\Delta s_R$ for the right wheel.  
Wait a second, we know how to estimate  $\Delta s_L$ and $\Delta s_R$ using the enconders!!!

<br><img src="../assets/ex7-imgs/kin-encoder2.png" style="width: 20%; height: auto;"></br>

Recall that $\Delta s_L = 2\pi R \frac{N_R}{N_{tot}}$ and  $\Delta s_R = 2\pi R \frac{N_R}{N_{tot}}$   
Where $N_L, N_R$, and $N_{tot}$ are the number of "ticks" counted by the left encoder, right encoder, and total number of ticks in the encoder.   
For instance, $N_L$ will count the number of ticks going from initial position to the next position of the MM.  

We can now derive $\Delta s$ and $\alpha \approx \Delta \theta$ from the values of  $\Delta s_L$ and $\Delta s_R$.  

$$\Delta s = \frac{\Delta s_L + \Delta s_R}{2}$$


$$\alpha \approx \Delta\theta = \frac{\Delta s_R - \Delta s_L}{2L}$$


Next, we will see how to calculate the new position based on the values of $\Delta s$ and $\Delta \theta$

<h2 style= "color:coral">Calculating New Position
</h2>

Recall that we need to calculate $\Delta x$, $\Delta y$, and $\Delta \theta$ in order to calculate a new position  
We already have a formula for $\Delta \theta$ üòÅ  

Assuming that the robot position estimates are calculated within small steps, we have the following approximation:

<br><img src="../assets/ex7-imgs/kin-calc-1.png" style="width: 20%; height: auto;"></br>

$$\Delta d \approx \Delta s$$

Now, if we know the angle $\sigma$ between $\Delta x$ and $\Delta d$ as shown in the image below, then we can use trigonometric to find $\Delta x$ and $\Delta y$

<br><img src="../assets/ex7-imgs/kin-calc-2.png" style="width: 20%; height: auto;"></br>

With the same assumption that the robot position estimates are calculated within small steps, we hate that 
$$\sigma = \frac{\Delta \theta}{2}$$

**Remark:** For a full description on these approximations see: <strong> <a href="https://medium.com/@nahmed3536/wheel-odometry-model-for-differential-drive-robotics-91b85a012299">Detailed Wheel Odometry for Differential Drive Robotics</a></strong>
</h6>

Finally, we can estimate the new position of the robot based on $\Delta s\approx \Delta d$ and $\sigma \approx \frac{\Delta \theta}{2}$

$$ \Delta x = \Delta s \cos(\theta + \frac{\Delta \theta}{2})$$
$$ \Delta y = \Delta s \sin(\theta + \frac{\Delta \theta}{2})$$
$$ \Delta x = x + \Delta x$$
$$ \Delta y = y + \Delta y$$
$$ \Delta \theta = \theta + \Delta \theta$$

Recall that: 
$$\Delta s = \frac{\Delta s_L + \Delta s_R}{2}$$
$$\Delta \theta  = \frac{\Delta s_R - \Delta s_L}{2L}$$

where $2L$ is the length of the robot in between wheels.  


**REMARK** In Ex7, you will implement the equations above to compute the positions of the robot given $\Delta s_L$ and $\Delta s_R$ (obtained from the topic dist_wheel)
The length $2L = 0.1$

<h2 style= "color:coral">Issues of Odometry
</h2>

- Errors accumulate (We made a few assumptions above that in the long run accumulate errors)
- Time steps: if too big, then some odometry will be missed
- Wheel slips: hard to model since it varies for each surface
