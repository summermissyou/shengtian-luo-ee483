<h1 style= "color:yellowgreen">Numerical methods
</h1>
</div>

- Dynamics and other robots operations are given in continuous equations forms, e.g., continuous on time
- Computers cannot deal with continuity. In other words, we need discretized forms of these equations
- Robots get updates in discrete-time. For instance, we can set the rate of your nodes.
- To build discrete but real time capable systems, we need: **numerical differentiation and integration**


<h2 style= "color:coral">Numerical Differentiation
</h2>

Derivative describe the instantenoues rate of change, e.g., instaneous velocity or acceleration.

$$f'(x)  = \lim_{h\rightarrow 0} \frac{f(x+h)-f(x)}{h}$$

*ISSUE*: Robots cannot take infinitesimal small steps. Updates are discrete

*SOLUTION*: Set $h$ to be a small value. The smaller the value the more accurate derivative value we get.

Velocity estimate example:

- Position (x,y) updated every $\frac{1}{10}$ of a second (10Hz)
- No velocity sensor in the robot. How to estimate velocity?
- Numerically differentiate 

<br><img src="../assets/ex7-imgs/numerical-vel-example1.png" style="width: 35%; height: auto;"></br>

$$v_x^*  = \frac{0.05}{0.1} = 0.5\ m/s$$

<br><img src="../assets/ex7-imgs/numerical-vel-example2.png" style="width: 50%; height: auto;"></br>

$$v_x^*  = \frac{d_2}{t_2-t_1} = 1\ m/s$$

Example for $x$ and $y$ velocity with position updated every $\delta_t$

$$v_x^*  = \frac{x_{current}-x_{previous}}{\delta_t} = \frac{\delta_x}{\delta_t} $$
$$v_y^*  = \frac{y_{current}-y_{previous}}{\delta_t} = \frac{\delta_y}{\delta_t}$$

<h2 style= "color:coral">Numerical Integration
</h2>

Intuitively, integrals compute the "area under the curve" 

**Riemann integral**
$$\int_a^b f(x) dx = \lim_{N\rightarrow \infty}\sum_{i=0}^{N}f(a+i\Delta_x)\Delta_x$$ 
where $\Delta_x = \frac{1}{N} (b-a)$

- No infinity values in computers ðŸ¥²
- What about finite step sizes? 
	- Multiple ways of choosing them

1. Assume everything is rectangles

$$\int_a^b f(x) dx \approx \sum_{i=0}^{N}f(a+i\Delta_x)\Delta_x$$ 
where $\Delta_x = \frac{1}{N} (b-a)$

**PRACTICAL**

- Usually time updates are quick
- We only need an estimate
- **Rectangular** method is usually sufficient
- If more precision is needed, then trapezoid or RK can be used.
